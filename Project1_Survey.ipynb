```python
import matplotlib.pyplot as plt
```

```python
for filename in upload.keys():
    data = pd.read_excel(filename)


ages = data["What is your age?"].tolist()

# Calculate mean, median, and mode
mean_age = np.mean(ages)
median_age = np.median(ages)
mode_age = stats.mode(ages).mode

# Create the box plot
plt.figure(figsize=(8, 6))
plt.boxplot(ages, vert=False)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.yticks([])  # Hide y-axis
plt.grid(True)


plt.scatter(mean_age, 1, color='red', label='Mean', zorder=5)
plt.scatter(median_age, 1, color='blue', label='Median', zorder=5)
plt.scatter(mode_age, 1, color='green', label='Mode', zorder=5)

plt.legend()

plt.show()
plt.show()
```
